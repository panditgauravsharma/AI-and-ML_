```python
##### Here we have two Kernel means two matrix of size 3*3, one for horizontal and second for vertical, both matrix will be
#added two get the final matrix and that matrix will be the edges matrix.
#  for first kernel it is x-axis and second is y-axis
```

```python

import cv2
import numpy as np
import matplotlib.pyplot as plt
img=cv2.imread("images/lenna.png",0)
```

```python
#sobel parametes: ksize, dx,dy,ddepth
sobelx=cv2.Sobel(img,cv2.CV_64F,1,0,ksize=3)
#Sobel for horizontal edges
sobely=cv2.Sobel(img,cv2.CV_64F,0,1,ksize=3)
#Sobel for vertical edges
#combine the sobel images optional
combined_sobel=cv2.add(sobelx,sobely)
```

```python
final_img=np.hstack([sobelx,sobely,combined_sobel])
cv2.imshow('All images',final_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,4,1)
plt.imshow(img,cmap='gray')
plt.title('original')
plt.subplot(1,4,2)
plt.imshow(sobelx,cmap='gray')
plt.title('binary threshold')
plt.subplot(1,4,3)
plt.imshow(sobely,cmap='gray')
plt.title('inverse binary threshold')
plt.subplot(1,4,4)
plt.imshow(combined_sobel,cmap='gray')
plt.title('Otsu image threshold')
```

##### Canny Edge Detection algorithm, it is for clear visibility for the edges that was not in using Sobel method

```python
edges=cv2.Canny(img,50,150)# adjust threshods if needed
#display results
cv2.imshow('Original Image',img)
cv2.imshow('Canny Edge Detection',edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.imshow(img,cmap='gray')
plt.title('original')
plt.subplot(1,2,2)
plt.imshow(edges,cmap='gray')
plt.title('Edge Detected by Canny')
```

```python

gray=cv2.imread('images/circle.png',0)


#Created a binary image where picels above 127 are set to 255 white
#and others to 0
thresh=cv2.threshold(gray,127,255,cv2.THRESH_BINARY)[1]
#cv2.findCountours finds contours in the binary image.
#cv2.RETR_TREE retrieves all contours and creates a hiercachy tree.
#cv2.CHAIN_APPROX_SIMPLE approxiamates contours to save memory.
contours, hierarchy=cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
#-1 DRAS ALL contours
#(0,255,0) is the color(green)
#2 is the thickness of the lines.
drawcont=cv2.drawContours(img,contours, -1,(0,255,0),2) # Draw all the contours in 
cv2.imshow('thresh',drawcont)


#Display results
cv2.imshow('Original Image',img)
cv2.imshow('Threshold_Image',thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
