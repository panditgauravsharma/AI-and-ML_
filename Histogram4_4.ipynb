##### THRESHOLDING,

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img=cv2.imread('images/lenna.png',0)
```

```python
img
```

```python
new_img=img>127 # white becomes white and black become black
```

```python
new_img
```

```python
plt.imshow(new_img,cmap='gray') # this is a boolean image
```

```python
inv_img=img<127 # true are those who are low intensity pixel so black becomes white and white becomes black
```

```python
plt.imshow(inv_img,cmap='gray')
```

```python
plt.imshow(img,cmap='gray')
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,3,1)
plt.imshow(img,cmap='gray')
plt.title('original')
plt.subplot(1,3,2)
plt.imshow(new_img,cmap='gray')
plt.title('binary')
plt.subplot(1,3,3)
plt.imshow(inv_img,cmap='gray')
plt.title('inverse binary')
```

```python
# Otsu's Thresholding , FG=I>TH, BG=I<TH(FOREGROUND AND BACKGROUND AND THRESHOLDING), it's threshold=highest variance,
#variance is variation amonng pixel, if TH=124 then above 124 all are white and below of 124 are black
```

```python
#Apply Otsu's thresholding
thresh_val, thresh_otsu=cv2.threshold(img,0,255,cv2.THRESH_OTSU)
```

```python
thresh_val
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,4,1)
plt.imshow(img,cmap='gray')
plt.title('original')
plt.subplot(1,4,2)
plt.imshow(new_img,cmap='gray')
plt.title('binary threshold')
plt.subplot(1,4,3)
plt.imshow(inv_img,cmap='gray')
plt.title('inverse binary threshold')
plt.subplot(1,4,4)
plt.imshow(thresh_otsu,cmap='gray')
plt.title('Otsu image threshold')
```
