```python
# 1. Traslate the image towrds right by 400 and up by 150 pixels
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image=cv2.imread('images/lenna.png')
height, width, dimension=image.shape
```

```python
tx=400
ty=-150
T=np.float32([[1,0,tx],[0,1,ty]])# create translate matrix using np.float/int32([[],[]])make 2d array float matrix by np
```

```python
image_translation=cv2.warpAffine(image,T,(width,height))# translate the image
im=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
img=cv2.cvtColor(image_translation,cv2.COLOR_BGR2RGB)# convert cv2 bgr to rgb for plt
```

```python
plt.imshow(img)
```

```python
plt.figure(figsize=(16,9))

plt.subplot(1,2,1)
plt.imshow(im)
plt.title('original image')

plt.subplot(1,2,2)
plt.imshow(img)
plt.title('translated image')
```

```python
#2. Rotate the image 60 degree anticlockwise
```

```python
center=(width//2,height//2)
rot=cv2.getRotationMatrix2D(center,60,1.0)# create matrix to mulitply
```

```python
trans=cv2.warpAffine(im,rot,(width,height))# create new matrix
```

```python
plt.imshow(trans)
```
