##### Dataset , Dimensional Reduction PCA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('datasets/wisc_bc_data.csv')
```

```python
x=df.drop(['id','diagnosis'],axis=1)
```

##### standardize the data

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_std=scaler.fit_transform(x)
```

```python
x_std.shape
```

##### Apply the PCA (Principle standard Analysis)

```python
from sklearn.decomposition import PCA
```

```python
# create the object
pca=PCA()
```

```python
x_pca=pca.fit_transform(x_std)# automatically used std, it uses eigen value and eigen vetor
```

```python
x_pca
```

```python
var=pca.explained_variance_ratio_
var
```

```python
x_pca.shape
```

```python
plt.figure(figsize=(16,9))
plt.grid()
plt.xlabel('Principal Components')
plt.ylabel('Variance')
plt.xticks(range(1,3))
plt.bar(range(1,31),var)
```

```python
# create the object
pca=PCA(n_components=10)# consider first 10 feature which are very important
```

```python
x_pca=pca.fit_transform(x_std)
```

```python
x_pca.shape
```

```python
plt.figure(figsize=(16,9))            # we are considering top 10 and this data is used to build ML algorithm
plt.grid()
plt.xlabel('Principal Components')
plt.ylabel('Variance')
plt.xticks(range(1,11))
plt.bar(range(1,11),pca.explained_variance_ratio_)
```
