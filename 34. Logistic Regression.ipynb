##### Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_scaled=scaler.fit_transform(x)   # we have to seperate this x_scaled
```

```python
pd.DataFrame(x_scaled).describe()
```

```python
# learn from here
```

##### Cross Validation

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test, y_train, y_test=train_test_split(x_scaled,y,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

##### Build the model

```python
from sklearn.linear_model import LogisticRegression # here we have to get how much 1 is matched and how much 0 is matched
```

```python
classifier=LogisticRegression() # this LogisticRegression() return the object to classifier to make them seperate
```

```python
# train the algorithm
classifier.fit(x_train,y_train) # now fit
```

##### Predict on the unseen data

```python
y_pred=classifier.predict(x_test)
```

```python
y_pred
```

```python
y_test.values
```

```python
y_test.value_counts()
```

##### Performance Evaluation

```python
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import accuracy_score, ConfusionMatrixDisplay
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
# 0-> matching 67 times and not matched 1, 1-> matching 22 times and not matched 10 times, first array for 0 and second array
# from second array, 
#Definition: A confusion matrix is a table that is used to evaluate the performance of a classification model. It presents a summary of the predictions
# made by the model against the actual outcomes (true labels) in a tabular format.
#Components:

#True Positive (TP): Cases where the model correctly predicts the positive class. mat
#True Negative (TN): Cases where the model correctly predicts the negative class.
#False Positive (FP): Cases where the model incorrectly predicts the positive class (Type I error).
#False Negative (FN): Cases where the model incorrectly predicts the negative class (Type II error).
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)# accuracy_score find the accuracy
```

```python
cm=confusion_matrix(y_test,y_pred)
```

```python
# using formula
cm.diagonal().sum()/cm.sum()
```

```python
print(classification_report(y_test,y_pred))
```

```python
#  for class 0 accuracy .99, from 68 there is 67 matched
# for class 1 accuracy .69, from 32 there is 22 mathced
# for class 1 accuracy must be 80% atleast so it is not a good model
# precision: 0->67/77, for 1-> 22/23
```

```python
# recall-0. accuracy of class 0/ sensitivity of algorithm that is recall

r0=67/68
r0
```

```python
# recall-1/ accuracy of class 1/ specificity of algorithm that is recall
r1=22/32
r1
```

```python
# precision-0/ Positive Predictive Value
p0=67/77
p0
```

```python
# precision-1/ Negative Predictive Value
p1=22/23
p1
```

```python
# f1-score -0
2*p0*r0/(p0+r0)# harmonic means
```

```python
# f1-score-1
2*p1*r1/(p1+r1)
```
