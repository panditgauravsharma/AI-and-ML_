```python
import cv2
#load the image
image=cv2.imread('images/lenna.png')
#flip the image horizontally
flipped_horizontal=cv2.flip(image, 0)
#flip the image vertically
flipped_vertical= cv2.flip(image, 1)
#flip the image diagonally
flipped_diagonal= cv2.flip(image,-1)
#show the flipped images
cv2.imshow('Original',image)
cv2.imshow('horizontal',flipped_horizontal)
cv2.imshow('vertical',flipped_vertical)
cv2.imshow('diagonal',flipped_diagonal)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
flip_hor=cv2.cvtColor(flipped_horizontal,cv2.COLOR_BGR2RGB)
flip_ver=cv2.cvtColor(flipped_vertical,cv2.COLOR_BGR2RGB)
flip_dia=cv2.cvtColor(flipped_diagonal,cv2.COLOR_BGR2RGB)
flip_org=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
import matplotlib.pyplot as plt
plt.subplot(1,4,1)
plt.imshow(flip_org)
plt.title('Original Image')
plt.subplot(1,4,2)
plt.imshow(flip_hor)
plt.title('horizontal image')#ulti image of original
plt.subplot(1,4,3)
plt.imshow(flip_ver)
plt.title('vertical image')#sidhi but face direction is change
plt.subplot(1,4,4)
plt.imshow(flip_dia)
plt.title('diagonal image')#ulti and face dircetion is change
```
