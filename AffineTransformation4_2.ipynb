##### Affine transformation is having all four- Scaling, Translation, Rotation, Shearing

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
img=cv2.imread("images/lenna.png")
#define transformation parameters
tx=50 #translation along x-axis
ty=100 # translation along y-axis
angle=50 #rotattion angle is 50 degree
scale=1.2 #scaling factor
shear_x=0.3 #Shearing along x-axis
shear_y=0.2 #shearing along y-axis
#create transformattion matrix
M=np.float32([[scale*np.cos(np.radians(angle)),shear_x,tx],[shear_y,scale*np.sin(np.radians(angle)),ty]])
#apply transformatin using warpAffine
img_affine=cv2.warpAffine(img,M,(512,512))
plt.subplot(1,2,1)
plt.imshow(img_affine)
plt.subplot(1,2,2)
plt.imshow(img)
#Display original and affine transformed image
cv2.imshow('Affine Transformed Image',img_affine)
#original image
cv2.imshow('original image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
