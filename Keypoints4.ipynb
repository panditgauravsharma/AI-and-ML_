```python
import cv2
import matplotlib.pyplot as plt
```

```python
image1=cv2.imread('images/eiphel1.png',0)
image2=cv2.imread('images/eiphel2.png',0)
cv2.imshow('df',image1)
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.imshow(image1,cmap='gray')
plt.subplot(1,2,2)
plt.imshow(image2,cmap='gray')
```

```python
sift=cv2.SIFT_create()
```

```python
keypoints_1, descriptors_1=sift.detectAndCompute(image1,None)
kyepoints_2, descriptors_2=sift.detectAndCompute(image2,None)
len(keypoints_1),len(kyepoints_2)
```

```python
#feature matching
bf=cv2.BFMatcher(cv2.NORM_L1)
matches=bf.match(descriptors_1,descriptors_2)
matches=sorted(matches,key=lambda x:x.distance)
img3=cv2.drawMatches(image1,keypoints_1,image2,kyepoints_2,matches[:50],image2, flags=2)
plt.imshow(img3)
```
