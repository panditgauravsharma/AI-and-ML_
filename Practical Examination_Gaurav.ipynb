```python
#q8. Implement the feature selection method chi2 and find top 5 feature
```

```python
df=pd.read_csv('datasets/grapes_new.csv')
df
```

```python
df = df.dropna()
```

```python
x=df.drop(['CLASS','SHAPE','SOIL_TYPE'],axis=1) #x=df.drop(['id','diagnosis'],axis=1)
y=df['CLASS']
```

```python
from sklearn.feature_selection import SelectKBest, chi2
```

```python
skf=SelectKBest(k=5, score_func=chi2)
```

```python
skf.fit(x,df['CLASS'])
```

```python
x.loc[:,skf.get_support()]
```

```python
#9. Apply feature scaling technique and tranform
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_scaled=scaler.fit_transform(x)
print(x_scaled)
```

```python
#10. Perform the cross validation with split ratio 80:20 and stratification
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train,  y_test=train_test_split(x_scaled,y,test_size=0.2,stratify=y,random_state=0)
```

```python
#11. Implement all classification algo to check the accuracy on the data.
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier

```

```python
clf_log=LogisticRegression()
clf_dt=DecisionTreeClassifier(random_state=0)
clf_rf=RandomForestClassifier(random_state=0,n_estimators=10)
clf_knn=KNeighborsClassifier(n_neighbors=5)
clf_svm=SVC(random_state=0,kernel='poly')
clf_nb=GaussianNB()
knn=KNeighborsClassifier(n_neighbors=4)
```

```python
clf_log.fit(x_train,y_train)
clf_dt.fit(x_train,y_train)
clf_rf.fit(x_train,y_train)
clf_knn.fit(x_train,y_train)
clf_svm.fit(x_train,y_train)
clf_nb.fit(x_train,y_train)
clf_knn.fit(x,y)
```

```python
y_pred_log=clf_log.predict(x_test)
y_pred_dt=clf_dt.predict(x_test)
y_pred_rf=clf_rf.predict(x_test)
y_pred_svm=clf_svm.predict(x_test)
y_pred_nb=clf_nb.predict(x_test)
y_pred_knn=clf_knn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print('Log: ',accuracy_score(y_test,y_pred_log))
print('DT: ',accuracy_score(y_test,y_pred_dt))
print('rf: ',accuracy_score(y_test,y_pred_rf))
print('svm: ',accuracy_score(y_test,y_pred_svm))
print('nb: ',accuracy_score(y_test,y_pred_nb))
print('knn: ',accuracy_score(y_test,y_pred_knn))
```

```python
# 12. Create the dataframe which shows the comparison of accuracy and speed of different classification algo implemented on this dataset
```

```python


# Creating DataFrame
df = pd.DataFrame({

'Log':  [accuracy_score(y_test,y_pred_log)],
'DT': [accuracy_score(y_test,y_pred_dt)],
'rf':  [accuracy_score(y_test,y_pred_rf)],
'svm':  [accuracy_score(y_test,y_pred_svm)],
'nb': [accuracy_score(y_test,y_pred_nb)],
'knn': [accuracy_score(y_test,y_pred_knn)],
})

print(df)
```
