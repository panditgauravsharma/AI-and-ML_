```python
import cv2
#load the image
image=cv2.imread('images/lenna.png')
#resizing the image to a fixed width and height
resized_image=cv2.resize(image,(100,200))#x,y
#show the resized image
cv2.imshow('resized image',resized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
#load the image
image=cv2.imread('images/lenna.png')
#resizing the image to a fixed width and height
resized_image=cv2.resize(image,None,fx=0.75,fy=0.75)#x,y
#show the resized image
cv2.imshow('resized image',resized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
#load the image
image=cv2.imread('images/lenna.png')
#resizing the image to a fixed width and height
resized_image=cv2.resize(image,None,fx=2.75,fy=1.75)#x,y
#show the resized image
cv2.imshow('resized image',resized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
from skimage.transform import resize
import matplotlib.pyplot as plt

#load the image
image=plt.imread('images/lenna.png')
resized_image=resize(image,(1000,1200), anti_aliasing=True)

#Resize with bicubic interpolation
#resized_image=resize(image,(1200,1300), anti_aliasing=True order=3)
#show the resized image
plt.subplot(1,2,1)
plt.imshow(image)
plt.subplot(1,2,2)
plt.imshow(resized_image)
```
